```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt


# Set the random seed as the last two digits of my roll number
np.random.seed(65)

# Create a 5x5 matrix A with random integers between 0 and 9
A = np.random.randint(0, 9, size=(5, 5))
print("Original matrix A:")
print(A)

# Convert the NumPy array to a Sympy Matrix
A_sympy = Matrix(A)

# Reduce matrix A to its Row Echelon Form
A_ref = A_sympy.rref()
A_rref = np.array(A_ref[0])
print("\nRow Echelon Form of matrix A:")
print(A_rref)
```

```python
# Perform Singular Value Decomposition (SVD)
U, Sigma, Vt = np.linalg.svd(A)
```

```python


# Print matrices U, Sigma, and V
print("Matrix U:")
print(U)
print("\nMatrix Sigma (Diagonal elements only):")
print(np.diag(Sigma))
print("\nMatrix V:")
print(Vt.T)  # Transpose Vt to get V

# Verify A = U * Sigma * Vt
reconstructed_A = np.dot(U, np.dot(np.diag(Sigma), Vt))
print("\nReconstructed A from U * Sigma * Vt:")
print(reconstructed_A)

# Rank 2 approximation
rank_2_approximation = np.dot(U[:, :2], np.dot(np.diag(Sigma[:2]), Vt[:2, :]))
print("\nRank 2 approximation of A:")
print(rank_2_approximation)

# Rank 3 approximation
rank_3_approximation = np.dot(U[:, :3], np.dot(np.diag(Sigma[:3]), Vt[:3, :]))
print("\nRank 3 approximation of A:")
print(rank_3_approximation)
```

```python


# Set the random seed as the last two digits of your roll number
np.random.seed(65)

# Generate random coefficients for a 5th degree polynomial
coefficients = np.random.randn(6)  # 6 coefficients for a 5th degree polynomial

# Define the polynomial function
def polynomial(x):
    return np.polyval(coefficients, x)

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Calculate y values using the polynomial function
y_values = polynomial(x_values)

# Plot the polynomial
plt.plot(x_values, y_values, label='Polynomial')

# Find maxima and minima
maxima = x_values[np.argmax(y_values)]
minima = x_values[np.argmin(y_values)]

# Annotate maxima and minima
plt.scatter(maxima, polynomial(maxima), color='red', label=f'Maxima: ({maxima:.2f}, {polynomial(maxima):.2f})')
plt.scatter(minima, polynomial(minima), color='green', label=f'Minima: ({minima:.2f}, {polynomial(minima):.2f})')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a 5th degree polynomial with Maxima and Minima')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
